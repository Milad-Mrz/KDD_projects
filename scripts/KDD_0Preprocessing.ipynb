#### Step 0: Visualizing and exploring data </br>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("../data/hour.csv")
print("number of instances: ", df.shape[0])

df6 = df.copy()  # Make a copy of the DataFrame to avoid modifying the original

# Updating attributes to original values
df6['atemp'] = df['atemp'] * (50 + 16) - 16
df6['temp'] = df['temp'] * (39 + 8) - 8
df6['hum'] = df['hum'] * 100
df6['windspeed'] = df['windspeed'] * 67



import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Visualizing usage of bike per month.
plt.figure(figsize=(15, 5))
sns.boxplot(x='mnth', y='cnt', data=df6, palette='Set2')
plt.title('Bike Rental Count per Month')
plt.xlabel('Month')
plt.ylabel('Count')
plt.show()
```

```python


# Visualizing usage of bike per hour.
plt.figure(figsize=(25, 8))
sns.boxplot(x='hr', y='cnt', data=df6, palette='Set2')
plt.title('Hourly Bike Rental Count')
plt.xlabel('Hour')
plt.ylabel('Count')
plt.show()
```

```python

# Visualizing impact of temperature on bike rental.
df7 = df6.copy()
df7['temp'] = df7['temp'].round()
plt.figure(figsize=(25, 10))
sns.boxplot(x='temp', y='cnt', data=df7, palette='Set2')
plt.title('Temperature vs Bike Rental')
plt.xlabel('Temperature')
plt.ylabel('Count')
plt.show()

```

```python
# Visualizing impact of temperature on bike rental.
df7 = df6.copy()
df7['hum'] = df7['hum']/100
df7['hum'] = df7['hum'].round(1)
plt.figure(figsize=(25, 10))
sns.boxplot(x='hum', y='cnt', data=df7, palette='Set2')
plt.title('Humidity vs Bike Rental')
plt.xlabel('Humidity')
plt.ylabel('Count')
plt.show()
```

```python
df7['windspeed'] = df7['windspeed'].round()
# Visualizing windspeed vs rental count of bike.
plt.figure(figsize=(20, 10))
sns.boxplot(x='windspeed', y='cnt', data=df7, palette='Set2')

#plt.scatter(df['windspeed'], df['cnt'], color='orange')
plt.title('Windspeed vs Rental Counts')
plt.xlabel('Windspeed')
plt.ylabel('Rental Counts')
plt.show()

```

```python

# Visualizing Weather situation vs bike rentals.
plt.figure(figsize=(10, 8))
sns.boxplot(x='weathersit', y='cnt', data=df6, palette='Set2')
plt.title('Weather vs Bike Rental')
plt.xlabel('Weather')
plt.xticks(np.arange(4), ('Clear', 'Cloudy', 'Light Shower', 'Heavy Shower' ))
plt.ylabel('Count')
plt.show()

```

```python

# Visualizing season vs bike rentals.
plt.figure(figsize=(10, 5))
sns.boxplot(x='season', y='cnt', data=df6, palette='Set2')
plt.title('Season vs Rental Counts')
plt.xlabel('Season')
plt.xticks(np.arange(4), ('1: Winter', '2: Spring', '3: Summer', '4: Fall'))
plt.ylabel('Rental Counts')
plt.show()


```
